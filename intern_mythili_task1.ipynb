# Student Performance Dataset

```python
# 1. Load Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/sample_data/student-mat.csv", sep=';')

# Display first 5 rows
df.head()
```

```python
# 2. Explore & Clean Data
# Dataset shape & data types
df.shape
df.dtypes

# Check missing values
df.isnull().sum()

# Remove duplicates
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
# 3. Analysis Questions
# 1. Average Final Grade (G3)
avg_g3 = df['G3'].mean()
avg_g3

# 2. How many students scored above 15?
above_15 = df[df['G3'] > 15].shape[0]
above_15

# 3. Is study time correlated with performance?
correlation = df['studytime'].corr(df['G3'])
correlation

# 4. Which gender performs better on average?
df.groupby('sex')['G3'].mean()
```

```python
# 4. Visualizations
# Histogram – Distribution of Final Grades
plt.figure()
plt.hist(df['G3'], bins=10)
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Frequency")
plt.show()

# Scatter Plot – Study Time vs Grades
plt.figure()
plt.scatter(df['studytime'], df['G3'])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()


# Bar Chart – Gender vs Average Score
gender_avg = df.groupby('sex')['G3'].mean()
plt.figure()
gender_avg.plot(kind='bar')
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```
